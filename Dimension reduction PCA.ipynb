```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import Normalizer
from sklearn.decomposition import PCA
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
path = 'D:\data science\seeds-width-vs-length.csv'
Column = ['width', 'length']
df = pd.read_csv(path, names = Column)
df.head()
```

dimention reduction works in 2 steps<br>
step 1: the dat is shifted axis and it normalizes the data with mean = 0 <br><t>
        The axis which is going to be shifted into the coordinate axis is called as its principal companents<br>
step 2: reduces the dimention"

```python
plt.scatter(df['width'], df['length'])
plt.axis('equal')
plt.show()
#correlaton of the data before performing PCA
correlation, pvalue = pearsonr(df['width'], df['length'])
print(correlation)
```

```python
model = PCA()
pca_features = model.fit_transform(df)
plt.scatter(pca_features[:,0], pca_features[:,1])
plt.axis('equal')
plt.show()
#correlation after PCA transformation
correlation, pvalue = pearsonr(pca_features[:,0], pca_features[:,1])
print(correlation)
```

We can clearly see that the correlation is of the order 10^-16

All this while we have de correlated the values and alligned it to the coordinate axis on the basis of its principal components
but now we would like to reduce the dimentions of the datset for easier analysis

```python
model = PCA()
model.fit(df)
features = range(model.n_components_)
plt.bar(features, model.explained_variance_)
plt.show()
```

As we can see that the first principal component has more varience than the other one, so we can work with a single component

```python
model = PCA(n_components = 1)
model.fit(df)
transformed_df = model.transform(df)
transformed_df.shape
```
